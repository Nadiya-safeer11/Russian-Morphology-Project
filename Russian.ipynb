```python
# Define a function to apply case suffixes to nouns
def get_case_forms(noun):
    case_forms = {}

    # Nominative case (no change)
    case_forms['nominative'] = noun

    # Genitive case
    if noun[-1] in 'а':
        case_forms['genitive'] = noun + 'а'  # Example rule for masculine nouns
    elif noun[-1] in 'я':
        case_forms['genitive'] = noun + 'и'
    else:
        case_forms['genitive'] = noun + 'а'

    # Dative case
    case_forms['dative'] = noun + 'у'  # Rule for masculine and neuter nouns

    # Accusative case
    case_forms['accusative'] = noun + 'а'  # For masculine nouns
    case_forms['accusative_plural'] = noun + 'ы'  # Plural form

    # Instrumental case
    case_forms['instrumental'] = noun + 'ом'  # Masculine/neuter
    case_forms['instrumental_plural'] = noun + 'ами'  # Plural form

    # Prepositional case
    case_forms['prepositional'] = noun + 'е'  # Singular
    case_forms['prepositional_plural'] = noun + 'ах'  # Plural form

    return case_forms


# Example usage
noun = "стол"
cases = get_case_forms(noun)

# Display the results
for case, form in cases.items():
    print(f"{case.capitalize()}: {form}")
```
